## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
plt.figure(figsize=(10, 6))
sns.histplot(df['ConvertedComp'], bins=10, kde=True)
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

plt.figure(figsize=(10, 6))
sns.boxplot(y=df['Age'])
plt.title('Boxplot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.figure(figsize=(10, 6))
bubble_plot = sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 200), data=df, legend=False, alpha=0.6)

# Adding labels and title
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs with Age as Bubble Size')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)

# Showing the plot
plt.show()
```

```python
# your code goes here
database_desire_next_year = df['DatabaseDesireNextYear'].dropna()

# Split the string values into separate entries
database_list = database_desire_next_year.str.split(';').explode()

# Get the top 5 databases
top_5_databases = database_list.value_counts().nlargest(5)

# Prepare data for pie chart
labels = top_5_databases.index
sizes = top_5_databases.values
explode = [0.1] + [0] * (len(labels) - 1)  # Explode the first slice

# Plotting the pie chart
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Showing the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Filter the data for the age group 30 to 35
age_filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate the median WorkWeekHrs and CodeRevHrs
median_work_week_hrs = age_filtered_df['WorkWeekHrs'].median()
median_code_rev_hrs = age_filtered_df['CodeRevHrs'].median()

# Prepare data for stacked bar chart
labels = ['Age 30-35']
work_week_hrs = [median_work_week_hrs]
code_rev_hrs = [median_code_rev_hrs]

# Plotting the stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(labels, work_week_hrs, label='Work Week Hours')
ax.bar(labels, code_rev_hrs, bottom=work_week_hrs, label='Code Review Hours')

# Adding labels and title
ax.set_ylabel('Median Hours')
ax.set_title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
ax.legend()

# Showing the plot
plt.show()
```
